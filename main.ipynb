```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
from matplotlib import pyplot as plt
plt.scatter(df.age,df.bought_insurance,marker = '+', color = 'red')
plt.ylabel('Age')
plt.xlabel('bought_insurance')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
# df[['age']] is a multidimensional array bcoz X has to be a multidim array
#  We will use 90% for training  and 10% for testing.
X_train, X_test, y_train, y_test = train_test_split(df[['age']],df.bought_insurance, test_size = 0.1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# Train the model
model.fit(X_train,y_train)
```

```python
X_test
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)  # the score is high because am using a very small dataset.
```

```python
model.predict_proba(X_test)
```
